```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.drop(columns = ['id','imdb_id','homepage','cast','tagline','overview','budget_adj'], inplace = True)
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df.dropna(how = 'any', subset = ['genres','director'],inplace =True)
```

```python
df.head(3)
```

```python
df.isnull().sum()
```

```python
df['production_companies'] = df['production_companies'].fillna(0)
df['keywords'] = df['keywords'].fillna(0)
```

```python
df.isnull().sum()
```

```python
df['popularity'] = df['popularity'].round(2)
```

```python
df.head()
```

```python
df.insert(3,'profit',df.revenue - df.budget) #.insert(where,name of new_col,value for new col)
df.head(3)
```

```python
df.insert(4,'roi',df.profit/df.budget)
```

```python
df.head()
```

```python
df['roi'] = df['roi'].round(2)
```

```python
df.head(3)
```

```python
df1 = df[['popularity','budget','revenue','profit','roi','vote_count','vote_average','release_year']]
```

```python
df1.isnull().sum()
```

```python
df.roi.value_counts()
```

```python
non_finite_values = ~np.isfinite(df['roi'])
```

```python
non_finite_values.sum()
```

```python
df['roi'] = df['roi'].replace([np.inf,-np.inf],np.nan)
```

```python
df3 = df.groupby('release_year')['popularity'].sum()
df3.plot(kind='line')
plt.title("Popularity amassed each Year")
plt.xlabel('Release Year')
plt.ylabel('Total Popularity')
```

```python
df4 = df.groupby('release_year')['vote_average'].mean()
df4.plot(kind='line')
plt.xlabel('Release Year')
plt.ylabel('Mean Rating')
```

```python
split = ['genres'] #list to contain values in genre column

for i in split:
    df[i] = df[i].apply(lambda x: x.split("|")) #lamba x:x.split splits each value and .apply() does it to every value in the column
df.head(3)
```

```python
df = df.explode('genres') #Expands the records for each genre
df.head(10)
```
